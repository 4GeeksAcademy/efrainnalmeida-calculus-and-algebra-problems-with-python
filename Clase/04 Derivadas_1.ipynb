```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Definir la función de desplazamiento y su derivada (velocidad)
def desplazamiento(t):
    resultado=np.sqrt(t)
    return (resultado)

def velocidad(t):
    return 1 / (2 * np.sqrt(t))
```

```python

# Crear datos de tiempo (de 1 a 100 segundos)
t = np.linspace(1, 100, 100)
y = desplazamiento(t)
v =velocidad(t)
```

```python
# Crear DataFrame para visualizar los valores
df = pd.DataFrame({'Tiempo (s)': t, 'Distancia (m)': y, 'Velocidad (m/s)': v})
```

```python
df
```

```python
# Graficar la función de desplazamiento y su derivada
plt.figure(figsize=(10, 5))

# Gráfica del desplazamiento
plt.subplot(1, 2, 1)
plt.plot(t, y, label=r'$y = \sqrt{t}$', color='blue')
plt.xlabel("Tiempo (s)")
plt.ylabel("Distancia (m)")
plt.title("Desplazamiento del carro")
plt.legend()
plt.grid()
```

```python
# Gráfica de la velocidad (derivada)
plt.subplot(1, 2, 2)
plt.plot(t, v, label=r'$\frac{1}{2\sqrt{t}}$', color='red')
plt.xlabel("Tiempo (s)")
plt.ylabel("Velocidad (m/s)")
plt.title("Velocidad del carro (derivada)")
plt.legend()
plt.grid()

plt.show()
```

```python
# Simulación del Teorema Central del Límite
# Tomamos 100 muestras aleatorias de la velocidad en diferentes tiempos
muestras = np.random.choice(v, size=(1000, 30))  # 1000 experimentos con 30 muestras cada uno
promedios = muestras.mean(axis=1)  # Promediamos las 30 muestras de cada experimento

# Graficar la distribución de los promedios
plt.figure(figsize=(7, 5))
plt.hist(promedios, bins=30, alpha=0.7, color='purple', edgecolor='black')
plt.xlabel("Velocidad Promedio (m/s)")
plt.ylabel("Frecuencia")
plt.title("Distribución de Promedios de Velocidad (TCL)")
plt.grid()
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Definir el tiempo en segundos (eje x) y la distancia en metros (f(x))
time = np.linspace(0, 10, 100)  # tiempo de 0 a 10 segundos
distance = np.sqrt(time)  # distancia en metros

# Crear la gráfica
fig = go.Figure()

# Agregar la función a la gráfica
fig.add_trace(go.Scatter(x=time, y=distance, mode='lines', name=r'$f(t) = \sqrt{t}$'))

# Títulos y etiquetas
fig.update_layout(
    title=r"Gráfica de $f(t) = \sqrt{t}$",
    xaxis_title="Tiempo (segundos)",
    yaxis_title="Distancia (metros)",
    template="plotly"
)

# Mostrar la gráfica
fig.show()
```
