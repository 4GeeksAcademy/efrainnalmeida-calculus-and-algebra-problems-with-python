## Vectores: 
#Son arreglos unidimensionales de números. Se utilizan para representar datos en Machine Learning, como características de un modelo o intensidades de # píxeles en una imagen.

```python
v1 = np.array([1, 2, 3])  # Vector de 3 elementos
print(v1)
```

## Matrices: 

#Son arreglos bidimensionales (filas y columnas). Se usan para representar imágenes, transformar datos o resolver ecuaciones lineales.

```python
M1 = np.array([ [1, 2], [3, 4] ])  # Matriz de 2x2
print(M1)
```

```python
# Vector de 10 unos
a1 = np.ones(10)
print(a1)
```

# Operaciones escalares en vectores y matrices: Se realizan operaciones básicas como suma, resta, multiplicación y división sobre cada elemento del vector o matriz.

```python
v2 = np.array([1, 2, 3, 4])
print(v2 + 2)  # Sumar 2 a cada elemento
print(v2 * 3)  # Multiplicar por 3
```

# Producto punto (Dot Product): Se usa en Machine Learning para cálculos de similitud, redes neuronales y regresión lineal.

```python
v3 = np.array([1, 2, 3])
v4 = np.array([4, 5, 6])
print(np.dot(v3, v4))  # Producto punto: (1*4 + 2*5 + 3*6)
```

```python
# Definir matrices
M2 = np.array([[1, 2, 3],
               [4, 5, 6]])

M3 = np.array([[7, 8],
               [9, 10],
               [11, 12]])

# Multiplicación de matrices
M_matrix = np.dot(M2, M3)  

print(M_matrix)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x): return x + 1
def g(x): return x**2 + 1

x = np.linspace(-2,2,100)
plt.plot(x,f(x),label='f(x) = x + 1')
plt.plot(x,g(x),label='g(x) = x² + 1')
plt.legend()
plt.show()
```

```python
def f(x):
    return x**2 - 2*x + 1

def df(x):
    h = 0.000001
    return (f(x + h) - f(x)) / h  # Aproximación de la derivada

x = np.linspace(-4, 4, 1000)
plt.plot(x, f(x), label='$f(x)$')
plt.plot(x, df(x), label="$f'(x)$")  # Derivada
plt.axhline(color='black')
plt.axvline(color='black')
plt.legend()
plt.show()
```
