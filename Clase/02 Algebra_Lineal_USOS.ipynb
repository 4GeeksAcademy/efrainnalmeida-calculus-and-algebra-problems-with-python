5.El método 'One hot encoding' codifica categorías.

```python
data = pd.DataFrame({'color': ['azul', 'rojo', 'verde', 'rojo']})
```

```python
data
```

```python
# Instanciar el codificador OneHotEncoder
encoder = OneHotEncoder(sparse_output=False) 
encoder
```

```python
# Ajustar y transformar los datos
encoded = encoder.fit_transform(data[['color']])
encoded
```

```python
# Crear un DataFrame con los resultados
encoded_df = pd.DataFrame(encoded, columns=encoder.categories_[0])
```

```python

# Mostrar el DataFrame con las columnas codificadas
print(encoded_df)
```
