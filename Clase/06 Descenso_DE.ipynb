```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Datos de entrada (x) y salida (y)
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])  # Relación y = 2x
```

```python
x
```

```python

# Inicialización del peso w
w = np.random.randn()
alpha = 0.01  # Tasa de aprendizaje o learning rate
num_epochs = 50  # Número de iteraciones

```

```python
# Para almacenar la evolución del peso y la pérdida
w_history = []
loss_history = []
```

```python
# Descenso del gradiente
for epoch in range(num_epochs):
    y_pred = w * x  # Predicción
    loss = np.mean((y_pred - y) ** 2)  # Cálculo del error cuadrático medio (MSE)
    
    grad_w = (2 / len(x)) * np.sum((y_pred - y) * x)  # Derivada del MSE respecto a w <<<<<-----------------------------------IMPORTANTE
    w -= alpha * grad_w  # Actualización del peso
    
    # Guardamos los valores para graficar
    w_history.append(w)
    loss_history.append(loss)


```

```python
# Graficar la evolución de w y la función de pérdida
plt.figure(figsize=(12, 5))

# Evolución de w
plt.subplot(1, 2, 1)
plt.plot(range(num_epochs), w_history, marker='o', label='w')
plt.xlabel('Épocas')
plt.ylabel('Valor de w')
plt.title('Evolución del Peso w')
plt.legend()

# Evolución de la función de pérdida
plt.subplot(1, 2, 2)
plt.plot(range(num_epochs), loss_history, marker='s', label='Loss')
plt.xlabel('Épocas')
plt.ylabel('Pérdida (MSE)')
plt.title('Evolución de la Función de Pérdida')
plt.legend()

plt.show()
```
