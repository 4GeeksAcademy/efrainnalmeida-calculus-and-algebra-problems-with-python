```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir la funci칩n de desplazamiento y su derivada
def desplazamiento(t):
    return np.sqrt(t)

def velocidad(t):
    return 1 / (2 * np.sqrt(t))
```

```python
# Crear datos para la curva
t = np.linspace(1, 100, 200)
y = desplazamiento(t)
```

```python
# Puntos donde queremos calcular las tangentes
tangentes_puntos = [1, 10, 50]
```

```python
type(tangentes_puntos)
```

```python
for i in tangentes_puntos:
    print(i)
```

```python
#La distancia (洧녽) est치 aumentando con el tiempo (洧노).
#El carro se sigue moviendo hacia adelante.
#La velocidad es positiva, lo que significa que el carro no est치 retrocediendo.
```

```python

# Graficar la funci칩n
plt.figure(figsize=(8, 5))
plt.plot(t, y, label=r'$y = \sqrt{t}$', color='blue')

# Dibujar las l칤neas tangentes en los puntos seleccionados
for t0 in tangentes_puntos:
    y0 = desplazamiento(t0)  # Distancia en t0
    v0 = velocidad(t0)  # Pendiente (velocidad) en t0
    
    # Definir la ecuaci칩n de la tangente: y_tan = y0 + v0 * (t - t0)

    #y = desplazamiento
    #
    t_tan = np.linspace(t0 - 10, t0 + 10, 50)  # Rango para dibujar la tangente
    y_tan = y0 + v0 * (t_tan - t0)  # Ecuaci칩n de la recta tangente
    
    # Graficar la l칤nea tangente
    plt.plot(t_tan, y_tan, '--', label=f'Tangente en t={t0}', alpha=0.7)

# Etiquetas y formato
plt.xlabel("Tiempo (s)")
plt.ylabel("Distancia (m)")
plt.title("Funci칩n y su derivada como l칤nea tangente")
plt.legend()
plt.grid()
plt.show()
```
